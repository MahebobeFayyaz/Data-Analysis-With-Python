## Data Aggregating and Grouping

```python
df_dummy1.head(2)
```

```python
grouped_mean = df_dummy1.groupby('Vehicle Type')['Booking Value'].mean()
grouped_mean
```

```python
df_dummy1.groupby(['Vehicle Type','Booking Status'])['Booking Value'].sum()
```

```python
# aggregrate
grouped_agg = df_dummy1.groupby('Vehicle Type')['Booking Value'].agg(['mean','sum','count'])
grouped_agg
```

```python
df1= pd.DataFrame({'key':['a','b','c'],'value1':[1,2,3]})
df2= pd.DataFrame({'key':['a','b','d'],'value1':[4,5,6]})
```

```python
## Merge DataFrame on the key columns
pd.merge(df1,df2, on ='key', how='inner')
```

```python
pd.merge(df1,df2, on ='key', how='outer')
```

```python
pd.merge(df1,df2, on ='key', how='left')
```

```python
pd.merge(df1,df2, on ='key', how='right')
```
